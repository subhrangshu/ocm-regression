```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
#import cartopy.crs as ccrs
```

```python
bio1 = xr.open_dataset("../Multi-site/Raw Data/Phytoplankton1.nc")
bio2 = xr.open_dataset("../Multi-site/Raw Data/Phytoplankton2.nc")
bio3 = xr.open_dataset("../Multi-site/Raw Data/Phytoplankton3.nc")
bio4 = xr.open_dataset("../Multi-site/Raw Data/Phytoplankton4.nc")
bio5 = xr.open_dataset("../Multi-site/Raw Data/Phytoplankton5.nc")
bio6 = xr.open_dataset("../Multi-site/Raw Data/Phytoplankton6.nc")

phys1 = xr.open_dataset("../Multi-site/Raw Data/Salinity1.nc")
phys2 = xr.open_dataset("../Multi-site/Raw Data/Salinity2.nc")
phys3 = xr.open_dataset("../Multi-site/Raw Data/Salinity3.nc")
phys4 = xr.open_dataset("../Multi-site/Raw Data/Salinity4.nc")
phys5 = xr.open_dataset("../Multi-site/Raw Data/Salinity5.nc")
phys6 = xr.open_dataset("../Multi-site/Raw Data/Salinity6.nc")
```

```python
def get_combo(lat,lon,bio,phys):
    bio_df = bio.sel(latitude=lat,longitude=lon).to_dataframe()
    bio_df.to_csv("Extracted Datasets/temp_bio_data.csv".format(lat,lon))
    bio_df = pd.read_csv("Extracted Datasets/temp_bio_data.csv".format(lat,lon))
    bio_df['time'] = bio_df.time.str.slice(0,7)
    bio_df = bio_df.drop(['depth'],axis=1)
    bio_df.to_csv("Extracted Datasets/temp_bio_data.csv".format(lat,lon), index=False)
    bio_df = pd.read_csv("Extracted Datasets/temp_bio_data.csv".format(lat,lon))

    phy_df = phys.sel(latitude=lat,longitude=lon).to_dataframe().fillna(0)
    phy_df.to_csv("Extracted Datasets/temp_phy_data.csv".format(lat,lon))
    phy_df = pd.read_csv("Extracted Datasets/temp_phy_data.csv".format(lat,lon))
    phy_df['time'] = phy_df.time.str.slice(0,7)
    phy_df = phy_df.drop(['depth'],axis=1)
    phy_df.to_csv("Extracted Datasets/temp_phy_data.csv".format(lat,lon), index=False)
    phy_df = pd.read_csv("Extracted Datasets/temp_phy_data.csv".format(lat,lon))
    
    return pd.merge(bio_df, phy_df, on=['time','latitude','longitude'])
```

```python
lat1 = list(range(int(bio1.latitude.values.min()), int(bio1.latitude.values.max()) + 1))
lat2 = list(range(int(bio2.latitude.values.min()), int(bio2.latitude.values.max()) + 1))
lat3 = list(range(int(bio3.latitude.values.min()), int(bio3.latitude.values.max()) + 1))
lat4 = list(range(int(bio4.latitude.values.min()), int(bio4.latitude.values.max()) + 1))
lat5 = list(range(int(bio5.latitude.values.min()), int(bio5.latitude.values.max()) + 1))
lat6 = list(range(int(bio6.latitude.values.min()), int(bio6.latitude.values.max()) + 1))

lon1 = list(range(int(bio1.longitude.values.min()), int(bio1.longitude.values.max()) + 1))
lon2 = list(range(int(bio2.longitude.values.min()), int(bio2.longitude.values.max()) + 1))
lon3 = list(range(int(bio3.longitude.values.min()), int(bio3.longitude.values.max()) + 1))
lon4 = list(range(int(bio4.longitude.values.min()), int(bio4.longitude.values.max()) + 1))
lon5 = list(range(int(bio5.longitude.values.min()), int(bio5.longitude.values.max()) + 1))
lon6 = list(range(int(bio6.longitude.values.min()), int(bio6.longitude.values.max()) + 1))
```

```python
l = []
m = []
for i in lat2:
    for j in lon2:
        l.append(get_combo(i,j, bio2, phys2))
m.append(pd.concat(l))
l = []
for i in lat3:
    for j in lon3:
        l.append(get_combo(i,j, bio3, phys3))
m.append(pd.concat(l))
l = []
for i in lat4:
    for j in lon4:
        l.append(get_combo(i,j, bio4, phys4))
m.append(pd.concat(l))
l = []
for i in lat5:
    for j in lon5:
        l.append(get_combo(i,j, bio5, phys5))
m.append(pd.concat(l))
l = []
for i in lat6:
    for j in lon6:
        l.append(get_combo(i,j, bio6, phys6))
m.append(pd.concat(l))
l = []
n = []
for i in lat1:
    for j in lon1:
        l.append(get_combo(i,j, bio1, phys1))
n.append(pd.concat(l))
```

```python
combined_data = pd.concat(m)
loc6_combined = pd.concat(n)
```

```python
combined_data['year'] = combined_data.time.str.slice(0,4)
combined_data['month'] = combined_data.time.str.slice(5,7)
combined_data.to_csv('Extracted Datasets/Combined_Data1.csv', index=False)


loc6_combined['year'] = loc6_combined.time.str.slice(0,4)
loc6_combined['month'] = loc6_combined.time.str.slice(5,7)
loc6_combined.to_csv('Extracted Datasets/loc1_combined.csv', index=False)
```
